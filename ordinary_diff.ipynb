```python
# the differential equation  dy/dx = x, at x=5, given y = 0 at x = 0
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt


```

```python
def diff_eqn1(y,x):
    dydx = x
    return dydx
```

```python
x=np.linspace(0,5,10)
```

```python
y=odeint(diff_eqn1,0,x)
```

```python
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y)
plt.grid()
plt.legend('y')
```

```python
#Second Order Differential Equation 
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sode(y,x):
    y,z=y
    dydx=[z,6*y-z]
    return dydx
```

```python
x=np.linspace(0,1,100)
```

```python
sol=odeint(sode,[5,-5],x)
```

```python
plt.plot(x,sol[:,0])
```

```python
yy=2*np.exp(2*x)+3*np.exp(-3*x)
```

```python
plt.plot(x,yy)
plt.plot(x,sol[:,0],"g")
```
